```python
import pandas as pd
df_BTC = pd.read_csv(r'D:\allen_trading_crypto\data\Price\BTCUSDT_UPERP_15m.csv')
df_ETH = pd.read_csv(r'D:\allen_trading_crypto\data\Price\ETHUSDT_UPERP_15m.csv')
```

```python
df_BTC.columns
```

```python
#set column names
df_BTC.columns = ['datetime', 'BTC_open', 'BTC_high', 'BTC_low', 'BTC_close', 'BTC_volume','BTC_price_change','BTC_price_pct_change','BTC_price_change_ratio']
df_ETH.columns = ['datetime', 'ETH_open', 'ETH_high', 'ETH_low',
                  
                  
                   'ETH_close', 'ETH_volume','ETH_price_change','ETH_price_pct_change','ETH_price_change_ratio']
```

```python
df_BTC.tail()
```

```python
#把BTC的datetime全部加2分鐘
df_BTC['datetime'] = pd.to_datetime(df_BTC['datetime']) + pd.Timedelta(minutes=2)
#轉回pd
df_BTC['datetime'] = df_BTC['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df_BTC.tail()
```

```python
df_ETH.tail()
```

```python
#merge dataframes
df = pd.merge(df_BTC, df_ETH, on='datetime', how='outer')
```

```python
df
```

```python
avg_diffenence = (df['BTC_close'] - df['ETH_close']).mean()
```

```python
#把df中的BTC_price_pct_change和ETH_price_pct_change shift(-1)
df['BTC_price_pct_change'] = df['BTC_price_pct_change'].shift(-1)
df['ETH_price_pct_change'] = df['ETH_price_pct_change'].shift(-1)
```

```python
#把avg_diffenence加到df中
df['avg_diffenence'] = avg_diffenence
```

```python
#針對每一個row 如果BTC_close 和 ETH_close 存在 且 如果BTC_close - ETH_clsoe > avg_diffenence 把 BTC_price_pct_change 和 ETH_price_pct_change中較大的數字*-1 另一個設為0
#else BTC_close - ETH_clsoe < avg_diffenence 把 BTC_price_pct_change 和 ETH_price_pct_change中較小的數字*1 另一個設為0

df['BTC_price_pct_change'] = df.apply(lambda x: x['BTC_price_pct_change'] if x['BTC_close'] - x['ETH_close'] > x['avg_diffenence'] else 0, axis=1)
df['ETH_price_pct_change'] = df.apply(lambda x: x['ETH_price_pct_change'] if x['BTC_close'] - x['ETH_close'] > x['avg_diffenence'] else 0, axis=1)

df
```
